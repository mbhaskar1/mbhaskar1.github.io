```python
f(x, y) = x^2 + y^2
g(x, y) = x^2 + y^2 + x*y - 1
```

```python
G = implicit_plot(g, (-2, 2), (-2, 2), color='red', gridlines='major', title='Plot of $g(x, y)$', figsize=[4,4])
G.show()
```

```python
F = contour_plot(f, (-2, 2), (-2, 2), cmap='winter', fill=False, 
                 gridlines='major', title='Contour Plot of $f(x, y)$', figsize=[4, 4])
F.show()
```

```python
grad = arrow((0, 1), (0, 2), color='orange')
gradx = arrow((0, 1), (-2/5, 6/5), color='purple', linestyle='dashed')
grady = arrow((0, 1), (2/5, 9/5), color='purple', linestyle='dashed')
GR = F+G+grad+gradx+grady
GR.show(title='Gradient of $f$ at point $P_1=(0, 1)$ *')
```

```python
grad2 = arrow((-1, 1), (-1 - sqrt(2)/2, 1 + sqrt(2)/2), color='orange')
GR2 = F+G+grad2
GR2.show(title='Gradient of $f$ at point $P_2=(-1, 1)$ *')
```
